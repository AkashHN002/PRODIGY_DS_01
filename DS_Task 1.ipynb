```python
df.dropna(inplace = True)
```

```python
categorical_cols = df.select_dtypes('object').columns
numerical_cols = df.select_dtypes(np.number).columns
```

```python
unique_ls = []
for col in categorical_cols:
    unique_ls.append([df[col].unique()])
pd.DataFrame(unique_ls, index = categorical_cols, columns = ['Unique values'])
```

## Distribution of Categorical variables

```python
df['gender'].value_counts().plot.bar(rot = 0)
plt.title('Distribution of gender')
plt.xlabel('Gender')
plt.show()
```

```python
df['work_type'].value_counts().plot.bar(rot = 0)
plt.title('Distribution of work type')
plt.xlabel('Work type')
plt.show()
```

```python
df['Residence_type'].value_counts().plot.bar(rot = 0)
plt.title('Distribution of Residence type')
plt.xlabel('Residence type')
plt.show()
```

```python
df['smoking_status'].value_counts().plot.bar(rot = 0)
plt.title('Distribution of Smoking status')
plt.xlabel('Smoking status')
plt.show()
```

## Distribution of Numerical variables

```python
df['age'].plot.hist(edgecolor = 'white', bins = 20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.show()
```

```python
df['heart_disease'].value_counts().plot.bar(rot = 0)
plt.title('Distribution of Heart disease')
plt.xlabel('Heart disease')
plt.show()
```

```python
gender_df = df.groupby('gender')['age'].mean()
gender_df.plot.bar(rot = 0)
plt.title('Average age by gender')
plt.xlabel('age')
plt.show()
```

```python
gender_df = df.groupby('gender')['avg_glucose_level'].mean()
gender_df.plot.bar(rot = 0)
plt.title('Average glucose level by gender')
plt.xlabel('Average glucose level')
plt.show()
```

```python
gender_df = df.groupby('gender')['body_mass_iindex'].mean()
gender_df.plot.bar(rot = 0)
plt.title('Average body mass index by gender')
plt.xlabel('Average body mass index')
plt.show()
```
